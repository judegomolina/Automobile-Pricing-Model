## Let's start Preprocessing

Let's first drop the repeated index column from the dataset.

```python
data = raw_data.drop(['Unnamed: 0'], axis=1)
```

```python
data.head()
```

Now we can focus on the second column of our dataset, "symboling", it represents the insurance risk level of a car. Symboling is a discrete variable usually taking values from -3 (pretty safe) to +3 (pretty risky).

```python
data['symboling'].unique()
```

As expected all values are integers and contained in the given range, so there's no further preprocessing for this column

Now the second column of our dataset is "normalized-losses", it is the relative average loss payment per insured vehicle year. This value is normalized for all autos within a particular size classification, and represents the average loss per car per year. The values normally range from 65 to 256.

```python
data['normalized-losses'].unique()
```

Checking the values we can see some "?" which represent missing values in this column that we have to deal with, so let's check what percentage of all the observations are missing values.

```python
sum(data['normalized-losses'] == '?') / data['normalized-losses'].shape[0]
```

As we can see only 20% of our observations conatin missing values for the "normalized-losses" feature, so we can just drop those observations.

```python
data = data.replace('?', np.nan)
data.dropna(subset=['normalized-losses'], axis=0, inplace=True)
sum(data['normalized-losses'] == '?')
```

```python
data.shape[0]
```
